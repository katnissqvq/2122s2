Customer Personality Analysis is a detailed analysis of a company’s ideal customers. It helps a business to better understand its customers and makes it easier for them to modify products according to the specific needs, behaviors and concerns of different types of customers.

Customer personality analysis helps a business to modify its product based on its target customers from different types of customer segments. For example, instead of spending money to market a new product to every customer in the company’s database, a company can analyze which customer segment is most likely to buy the product and then market the product only on that particular segment.

## 1.Load the dataset, check all column types and check for missing values. Drop rows with missing values in the Education column.

* Use pandas to load the csv file, using ";" as seperation mark for the data.

```python
import pandas as pd
marketing_campaign=pd.read_csv('marketing_campaign.csv',sep=';')
marketing_campaign
```

* Print all the column names and their types with the function .dtypes

```python
marketing_campaign.dtypes
```

* Check missing values in the dataframe with the function .isnull().sum(), and aquire that Education column contains two rows with missing value

```python
marketing_campaign.isnull().sum()
```

* Use the .dropna()function to drop the rows with missing values, specify the code execution under the required column 'Education'. Assign the revised dataframe a name.

```python
marketing2=marketing_campaign.dropna(subset=['Education'])
```

* Re-confirm the successful drop of rows by using the isnull().sum() function again, ensuring that there is no missing values for 'Education'.

```python
marketing2.isnull().sum()
```

## 2.Create a smaller dataset consisting of households where the MntWines variable is lower than 160 and where MntMeatProducts is higher than 40.

* Create the new dataset "small" by filtering the original dataframe with the conditional requirement for columns. Only the rows whose 'MntWines' column and 'MntMeatProduct' column met the required condition would be selected.

```python
small=marketing2[(marketing2['MntWines']<160) & (marketing2['MntMeatProducts']>40)]
small
```

```python
small.isnull().sum()
```

## 3.What is the mean number of teenagers in these households in your smaller dataset?

* Use the .mean function to calculate the mean number of column 'Teenhome' in dataframe 'small'

```python
small['Teenhome'].mean()
```

## 4.What are the most and least common education levels in your smaller dataset?
